```
try:
  order = []
  def Top(qty):
    if(0<=qty<=2):
      price = 400*qty
    else:
      price = 350*qty
    print('Total price ='+str(price))
    p = int(input('Would you like to proceed? 1 for Yes, any other character for No\n'))
    if(p==1):
      tops = ['Tops', price]
      order.append(tops)
      takeorder()
  def Tshirt(qty):
    if(0<=qty<=2):
      price = 250*qty
    else:
      price = 200*qty
    print('Total price ='+str(price))
    p = int(input('Would you like to proceed? 1 for Yes, any other character for No\n'))
    if(p==1):
      tshirts = ['T-shirts', price]
      order.append(tshirts)
      takeorder()
  def Skirt(qty):
    if(0<=qty<=2):
      price = 600*qty
    else:
      price = 500*qty
    print('Total price ='+str(price))
    p = int(input('Would you like to proceed? 1 for Yes, any other character for No\n'))
    if(p==1):
      skirts = ['Skirts', price]
      order.append(skirts)
      takeorder()
  def Jeans(qty):
    if(0<=qty<=2):
      price = 1250*qty
    else:
      price = 1000*qty
    print('Total price ='+str(price))
    p = int(input('Would you like to proceed? 1 for Yes, any other character for No\n'))
    if(p==1):
      jeans = ['Jeans', price]
      order.append(jeans)
      takeorder()
  def Jacket(qty):
    if(0<=qty<=2):
      price = 750*qty
    else:
      price = 700*qty
    print('Total price ='+str(price))
    p = int(input('Would you like to proceed? 1 for Yes, any other character for No\n'))
    if(p==1):
      jackets = ['Jackets', price]
      order.append(jackets)
      takeorder()
  def Dress(qty):
    if(0<=qty<=2):
      price = 750*qty
    else:
      price = 700*qty
    print('Total price ='+str(price))
    p = int(input('Would you like to proceed? 1 for Yes, any other character for No\n'))
    if(p==1):
      dresses = ['Dresses', price]
      order.append(dresses)
      takeorder()
  def Shoes(qty):
    if(0<=qty<=2):
      price = 2250*qty
    else:
      price = 2000*qty
    print('Total price ='+str(price))
    p = int(input('Would you like to proceed? 1 for Yes, any other character for No\n'))
    if(p==1):
      shoes = ['Shoes', price]
      order.append(shoes)
      takeorder()
  def order_summary():
    tot = 0
    print('Order summary: '+ str(order))
    for i in range(len(order)):
      tot = tot+ order[i][1]
    print("Order total: Rs."+str(tot))
    print('Thanks for shopping!')
    f = open("apparel.txt",'w')
    for i in range(len(order)):
      f.write(str(order[i]))
  def takeorder():
    apparel = {1:'Top', 2: 'T-shirt', 3:'Skirt', 4:'Jeans', 5:'Jacket', 6:'Dress', 7:'Shoes'}
    choice = int(input('Choose what you want to buy: \n1.Top\n2.T-Shirt\n3.Skirt\n4.Jeans\n5.Jacket\n6.Dress\n7.Shoes\nPress 0 to Exit\n'))  
    if(choice!=0):
      if(apparel[choice] == 'Top'):
        q = int(input('Enter qty of tops you would like to buy: '))
        Top(q)
      elif(apparel[choice] == 'T-shirt'):
        q = int(input('Enter qty of T-shirts you would like to buy: '))
        Tshirt(q)
      elif(apparel[choice] == 'Skirt'):
        q = int(input('Enter qty of skirts you would like to buy: '))
        Skirt(q)
      elif(apparel[choice] == 'Jeans'):
        q = int(input('Enter qty of jeans you would like to buy: '))
        Jeans(q)
      elif(apparel[choice] == 'Jacket'):
        q = int(input('Enter qty of jackets you would like to buy: '))
        Jacket(q)
      elif(apparel[choice] == 'Dress'):
        q = int(input('Enter qty of dresses you would like to buy: '))
        Dress(q)
      elif(apparel[choice] == 'Shoes'):
        q = int(input('Enter qty of pairs of shoes you would like to buy: '))
        Shoes(q)
      else:
        print('Invalid option!')
        takeorder()
    else:
      order_summary()
  takeorder()
except (ValueError, KeyError):
  print('Please select a valid option(Must be a number)!\n')
  takeorder()


```
